## Question 1:
Evaluate derivative with respect to $x$:
$$
f(x) = x^4e^{\left(2x+a\right)^2}
$$


Product rule:
$$(fg)' = f'g + fg'$$
Chain rule:
$$f'(x) = h'(g(x)) \cdot g'(x)$$
Work:
$$f'(x) = 4x^3 e^{(2x+a)^2} + x^4 \cdot e^{(2x+a)^2} \cdot 2(2x+a)^{2-1} \cdot 2$$
$$f'(x) = 4x^3 e^{(2x+a)^2} + x^4 \cdot e^{(2x+a)^2} \cdot 4(2x+a)$$

$$= 4x^3 e^{(2x+a)^2}\left[1 + x(2x+a)\right]$$

## Question 2:
Evaluate the following integral:
$$
\int_{-1}^{2} 5t^4\;dt
$$


$$\int_{-1}^{2} 5t^4\;dt = \left[t^5\right]_{-1}^{2} = 2^5 - (-1)^5 = 32 - (-1) = 33$$

## Question 3:
If, $\hat{A} = \hat{A}^\dagger$, what kind of operator is $\hat{A}$?

$\hat{A}$ is a **Hermitian** operator.

---
## Question 4:
If, $\hat{A}\hat{A}^\dagger=\hat{A}^\dagger\hat{A}=\hat{I}$, what kind of operator is $\hat{A}$?

$\hat{A}$ is a **Unitary** operator.

---
## Question 5:
If, $\hat{A}\hat{A}=\hat{A}$, what kind of operator is $\hat{A}$?

$\hat{A}$ is an **Idempotent** operator.

---
## Question 6:
If $\hat{A}$ is Idempotent and Involutory, what is $\hat{A}$?

Idempotent gives $\hat{A}\hat{A} = \hat{A}$, and Involutory gives $\hat{A}\hat{A} = \hat{I}$. Therefore $\hat{A} = \hat{I}$.

## Question 8:
 Given the following :
- $\mathbf{x} = \begin{pmatrix}a\\b\end{pmatrix}$
- $\mathbf{y} = \begin{pmatrix}c\\d\end{pmatrix}$
- $\mathbf{X} = \begin{pmatrix}a & b \\ c & d\end{pmatrix}$
- $\mathbf{Y} = \begin{pmatrix}c & d \\ e & f\end{pmatrix}$

Compute: 
1. $\mathbf{Y}\mathbf{x} = ?$
1. $\mathbf{x}^\dagger\mathbf{Y} = ?$
1. $\mathbf{X}^\dagger\mathbf{Y} = ?$ 
1. $Trace(\hat{X}) = ?$

1. $\mathbf{Y}\mathbf{x} = \begin{pmatrix}c & d \\ e & f\end{pmatrix}\begin{pmatrix}a\\b\end{pmatrix} = \begin{pmatrix}ca+db \\ ea+fb\end{pmatrix}$

2. $\mathbf{x}^\dagger\mathbf{Y} = \begin{pmatrix}a^* & b^*\end{pmatrix}\begin{pmatrix}c & d \\ e & f\end{pmatrix} = \begin{pmatrix}a^*c+b^*e & a^*d+b^*f\end{pmatrix}$

3. $\mathbf{X}^\dagger\mathbf{Y} = \begin{pmatrix}a^* & c^* \\ b^* & d^*\end{pmatrix}\begin{pmatrix}c & d \\ e & f\end{pmatrix} = \begin{pmatrix}a^*c+c^*e & a^*d+c^*f \\ b^*c+d^*e & b^*d+d^*f\end{pmatrix}$

4. $\text{Trace}(\hat{X}) = a + d$

## Question 9:
Compute the following commutator: $\left[\begin{pmatrix} 1 & 2 \\ 3 & 4\end{pmatrix},\begin{pmatrix} 5 & 6 \\ 7 & 8\end{pmatrix}\right]$

where $\left[A,B\right] = AB - BA$, and the multiplication operation is the usual matrix multiplication.

$$AB = \begin{pmatrix}1\cdot5+2\cdot7 & 1\cdot6+2\cdot8 \\ 3\cdot5+4\cdot7 & 3\cdot6+4\cdot8\end{pmatrix} = \begin{pmatrix}19 & 22 \\ 43 & 50\end{pmatrix}$$

$$BA = \begin{pmatrix}5\cdot1+6\cdot3 & 5\cdot2+6\cdot4 \\ 7\cdot1+8\cdot3 & 7\cdot2+8\cdot4\end{pmatrix} = \begin{pmatrix}23 & 34 \\ 31 & 46\end{pmatrix}$$

$$[A,B] = AB - BA = \begin{pmatrix}19-23 & 22-34 \\ 43-31 & 50-46\end{pmatrix} = \begin{pmatrix}-4 & -12 \\ 12 & 4\end{pmatrix}$$

## Question 10:
Define the four Pauli matrices $I, X, Y, Z$: 
$I = \begin{pmatrix} 1 & 0 \\ 0 & 1\end{pmatrix}, X = \begin{pmatrix} 0 & 1 \\ 1 & 0\end{pmatrix}, Y = \begin{pmatrix} 0 & -i \\ i & 0\end{pmatrix}, Z = \begin{pmatrix} 1 & 0 \\ 0 & -1\end{pmatrix}$.

Define $|0\rangle = \begin{pmatrix} 1 \\ 0\end{pmatrix}, |1\rangle = \begin{pmatrix} 0 \\ 1\end{pmatrix}$.

Compute the expectation values: 
1. $\langle 0|X|0\rangle$, 
2. $\langle 1|X|1\rangle$,
3. $\langle 0|Y|0\rangle$,
4. $\langle 1|Y|1\rangle$,
5. $\langle 0|Z|0\rangle$,
6. $\langle 1|Z|1\rangle$.

1. $\langle 0|X|0\rangle = \begin{pmatrix}1 & 0\end{pmatrix}\begin{pmatrix}0\\1\end{pmatrix} = 0$
2. $\langle 1|X|1\rangle = \begin{pmatrix}0 & 1\end{pmatrix}\begin{pmatrix}1\\0\end{pmatrix} = 0$
3. $\langle 0|Y|0\rangle = \begin{pmatrix}1 & 0\end{pmatrix}\begin{pmatrix}0\\i\end{pmatrix} = 0$
4. $\langle 1|Y|1\rangle = \begin{pmatrix}0 & 1\end{pmatrix}\begin{pmatrix}-i\\0\end{pmatrix} = 0$
5. $\langle 0|Z|0\rangle = \begin{pmatrix}1 & 0\end{pmatrix}\begin{pmatrix}1\\0\end{pmatrix} = 1$
6. $\langle 1|Z|1\rangle = \begin{pmatrix}0 & 1\end{pmatrix}\begin{pmatrix}0\\-1\end{pmatrix} = -1$

## Question 11:
Define the four Pauli matrices $I, X, Y, Z$ as well as the basis states $|0\rangle, |1\rangle$ as in the previous question.

Define $|\psi\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)$.

Compute the expectation values: 
1. $\langle \psi|X|\psi\rangle$, 
2. $\langle \psi|Y|\psi\rangle$.

$|\psi\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix}1\\1\end{pmatrix}$, so $\langle\psi| = \frac{1}{\sqrt{2}}\begin{pmatrix}1 & 1\end{pmatrix}$

1. $\langle\psi|X|\psi\rangle$: $\quad X|\psi\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix}1\\1\end{pmatrix}$, so $\langle\psi|X|\psi\rangle = \frac{1}{2}(1+1) = \mathbf{1}$

2. $\langle\psi|Y|\psi\rangle$: $\quad Y|\psi\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix}-i\\i\end{pmatrix}$, so $\langle\psi|Y|\psi\rangle = \frac{1}{2}(-i+i) = \mathbf{0}$
